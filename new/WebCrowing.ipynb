#### for문 추가

```python
# 문서가 저장되고 있는 위치 변수
download_dir = "C:/Users/admin/Downloads"

# chrome 옵션 설정
options = Options()
options.add_argument(download_dir) # 다운로드 경로 설정

# 드라이버 불러오기
driver = webdriver.Chrome(options=options)
# 페이지 열기
driver.get('https://www.mrtc.re.kr/data/05.php')
time.sleep(3)

# 초기 element 값 설정
brief_element = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[3]/table/tbody/tr[1]/td[3]/a')
hrep = brief_element.get_attribute('href')

# 페이지 초기값
page_num = 4

# while문을 이용하여 전체 페이지 순회
while page_num <= 13:
    print(f'페이지 {page_num}')
    # for문을 이용하여 전체 페이지 문서 다운로드
    for i in range(10):
        brief_element = driver.find_elements(By.CLASS_NAME, 'subject > a')
        num = f"{page_num-4}{i+1}"
        # 총 문서 숫자 + 1
        p_num = 48-int(num)
        print(f'문서 {p_num}')
        hrep = brief_element[i].get_attribute('href')
        print(hrep)
        # 페이지 열기 
        driver.get(hrep)
        time.sleep(3)
        
        try:
            download_element = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[2]/ul/li[1]/a')
            # 다운로드 하기
            download_element.click()
            print(download_element)
            time.sleep(3)
            
        except:
            print("다운로드 링크가 없습니다.")
            
        # 다운로드 완료
        
        # 다운로드 창 자동 닫기
        all_windows = driver.window_handles
        
        # 원래 창 저장하기
        original_window = driver.current_window_handle
        
        for window in all_windows[1:]:
            driver.switch_to.window(window)
            driver.close()
            
        # 다운로드 파일 이름 바꾸기
        download_files = os.listdir(download_dir)
        download_files.sort(key=lambda x:os.path.getmtime(os.path.join(download_dir, x)), reverse=True)
        latest_file = download_files[0]
        
        # 파일 이름 변경
        new_file_name = f"{p_num}_"+latest_file
        os.rename(os.path.join(download_dir, latest_file), os.path.join(download_dir, new_file_name))
            
        # switch 했던 창 다시 원상복구 하기
        driver.switch_to.window(original_window)
        print("현재 URL:", driver.current_url)

        # 페이지 닫기
        driver.back()
        time.sleep(3)
        
    # 다음 페이지로 이동
    next_page = driver.find_element(By.XPATH, f'//*[@id="inner_wrap"]/div[2]/div[4]/a[{page_num}]')
    next_page.click()
    time.sleep(3)
    
    page_num += 1
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import os
import time

# 문서가 저장되고 있는 위치 변수
download_dir = "C:/Users/admin/Downloads"

# chrome 옵션 설정
options = Options()
# prefs = {'download.default_directory': download_dir}
# options.add_experimental_option('prefs', prefs)

# 드라이버 불러오기
driver = webdriver.Chrome(options=options)

# 페이지 열기
driver.get('https://www.mrtc.re.kr/data/08.php')
time.sleep(3)

# 페이지 초기값
page_num = 4

# while문을 이용하여 전체 페이지 순회
while page_num <= 13:
    print(f'페이지 {page_num}')

    # 1. 문서 리스트 가져오기
    brief_elements = driver.find_elements(By.CSS_SELECTOR, '.subject > a')
    href_list = [elem.get_attribute('href') for elem in brief_elements]

    # 2. 문서 하나씩 열기
    for i in range(len(href_list)):
        num = f"{page_num-4}{i+1}"
        p_num = 48 - int(num)

        print(f'문서 {p_num}')
        print(i)

        hrep = href_list[i]
        print(hrep)

        # 문서 페이지 열기
        driver.get(hrep)
        time.sleep(3)

        try:
            # 다운로드 요소 찾기
            element = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[2]/ul/li[3]/a')
            print(f"{p_num} 번째 문서:", element.text)

            # 다운로드하기
            element.click()
            time.sleep(3) # 다운로드 대기

            # 다운로드 창 자동 닫기
            all_windows = driver.window_handles
            original_window = driver.current_window_handle

            if len(all_windows) > 1:
                for window in all_windows[1:]:
                    driver.switch_to.window(window)
                    driver.close()

            driver.switch_to.window(original_window)

            # 다운로드 파일 이름 바꾸기
            time.sleep(2) # 파일 생성 기다림
            download_files = os.listdir(download_dir)
            download_files.sort(key=lambda x: os.path.getmtime(os.path.join(download_dir, x)), reverse=True)
            latest_file = download_files[0]

            new_file_name = f"{p_num}_"+latest_file
            os.rename(os.path.join(download_dir, latest_file), os.path.join(download_dir, new_file_name))

            # 원래 리스트 페이지로 돌아가기
            driver.back()
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.ID, "inner_wrap"))
            )

        except NoSuchElementException:
            print("다운로드할 요소 없음")
            driver.back()
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.ID, "inner_wrap"))
            )
            continue

        # 3. 다음 페이지로 이동
    try:
        driver.get('https://www.mrtc.re.kr/data/08.php')
        time.sleep(3)

        next_page = driver.find_element(By.XPATH, f'//*[@id="inner_wrap"]/div[2]/div[4]/a[{page_num}]')
        next_page.click()
        time.sleep(3)

    except Exception as e:
        print(f"다음 페이지 버튼 못 찾음: {e}")
        break

    page_num += 1

    # 모든 작업 끝나면 브라우저 종료
driver.quit()
```
