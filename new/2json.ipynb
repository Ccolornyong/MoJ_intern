### Treemap에서 사용할 exel을 json으로 변환

```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv("C:/HANSSAK/SecureGate/download/분석본_250528_장기체류(등록+거소)외국인_데이터셋(2015_2024, 202501_04).csv", encoding='cp949')
df
```

```python
result = []

for sido, group in df.groupby('시도명'):
    item = []
    for _, row in group.iterrows():
        item.append({
            "자격_대분류": row['자격_대분류'],
            "외국인수": row['외국인수']
        })
        
    result.append({
        "시도명": sido,
        "data": item
    })
    
    
result
```

```python
with open("C:/HANSSAK/SecureGate/download/분석본_250528_장기체류(등록거소)_전처리.json", "w", encoding='utf-8') as f:
    json.dump(result, f, ensure_ascii=False, indent=2)
```

### excel에 자격_코드명 특성 추가

```python
kor_title = pd.read_excel("C:/HANSSAK/SecureGate/download/250529_대분류자격코드명_메타.xlsx")
kor_title
```

```python
kor_dict = dict(zip(kor_title['자격_대분류'], kor_title['LW_NM']))
kor_dict
```

```python
# 기존 df에 자격_코드명 특성 추가 후 map으로 연결
df['자격_코드명'] = df['자격_대분류'].map(kor_dict)
df
```

```python
df.loc[df['자격_대분류'] == '0', '자격_코드명'] = '자격없음'
df
```

```python
df.to_csv("C:/HANSSAK/SecureGate/download/코드명추가_202412_장기체류(등록거소)_전처리.csv", index=False, encoding='cp949')
df
```
