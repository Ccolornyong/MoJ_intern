#### 2504to2505 추가

```python
# 안전한 계산을 위해 0으로 나누는 경우 대비
def calc_rate_change(new, old):
    return (new - old) / old if old != 0 else None

# 증감률 계산
df_gg_grouped["2504to2505"] = (((df_gg_grouped["202505_외국인수"] - df_gg_grouped["202504_외국인수"]) / df_gg_grouped["202504_외국인수"]) * 100).round(2)

df_gg_grouped
```
