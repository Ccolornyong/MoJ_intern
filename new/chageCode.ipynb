```python
## import
import pandas as pd
import numpy as np
```

```python
# 원 데이터 불러와서 code, label dict 만들기
code_df = pd.read_csv("C:/Users/admin/workspace/WorldCountry.csv")
code_df
```

```python
code_dict = code_df.set_index('@label')['@code'].to_dict()
code_dict
```

```python
## code 바꾸기
final_df = pd.read_csv("C:/Users/admin/workspace/출입국자_지도_국가코드_6대륙_final.csv")
final_df
```

```python
final_df['NAT_HAN_NM'].to_list()
```

```python
for k, v in code_dict.items():
    if k not in final_df['NAT_HAN_NM'].values:
        print(f"{k} is not in final_df")
```

```python
noname_dict = {
    '기아나': '가이아나',                 # 유사
    '슬로바키아': '슬로바크',            # 음역 차이
    '보스니아헤르체고비나': '보스니아-헤르체고비나',  # 하이픈 차이
    '마케도니아': '북마케도니아',         # 정식 명칭 반영
    '서사하라': None,                    # 리스트에 없음
    '차드': '차트',                      # 표기 차이 (Chad)
    '스와질란드': '에스와티니',           # 국가명 변경됨
    '부루키나파소': '부르키나파소',       # 철자 오류 수정
    '남수단': '남수단공화국',             # 전체 이름 사용
    '대한민국': '한국',                   # 일상명칭과 공식명칭
    '예멘': '예멘공화국',                 # 전체 이름 사용
    '아랍에미리트': '아랍에미리트연합',     # 전체 이름 사용
    '키르기스스탄': '키르기즈',           # 줄여 쓴 이름
    '러시아': '러시아(연방)'              # 전체 이름 사용
}

noname_dict
```

```python
reverse_dict = {v: k for k, v in noname_dict.items() if v is not None}
```

```python
# final_df의 이름 바꾸기

final_df['NAT_HAN_NM'] = final_df['NAT_HAN_NM'].map(lambda x: reverse_dict.get(x, x))
final_df
```

```python
for k, v in code_dict.items():
    if k in final_df['NAT_HAN_NM'].values:
        final_df.loc[final_df['NAT_HAN_NM'] == k, 'country_code'] = v
        print(f"{k} is in final_df and replaced with {v}")
```
