```python
# 파일 목록과 연도 매핑
files = {
    "2020": "C:/HANSSAK/SecureGate/Download/2020_250529_장기체류(등록+거소)외국인_데이터셋.csv",
    "2021": "C:/HANSSAK/SecureGate/Download/2021_250529_장기체류(등록+거소)외국인_데이터셋.csv",
    "2022": "C:/HANSSAK/SecureGate/Download/2022_250529_장기체류(등록+거소)외국인_데이터셋.csv",
    "2023": "C:/HANSSAK/SecureGate/Download/2023_250529_장기체류(등록+거소)외국인_데이터셋.csv",
    "2024": "C:/HANSSAK/SecureGate/Download/2024_250529_장기체류(등록+거소)외국인_데이터셋.csv"
}

df_list = []

for year, file in files.items():
    df = pd.read_csv(file, encoding='cp949')

    # 필요한 컬럼만 가져오고, 컬럼 이름 변경
    df = df.groupby(["시도명", "자격_대분류", "자격_코드명"], as_index=False).sum()
    df.rename(columns={"외국인수": f"{year}_외국인수"}, inplace=True)

    df_list.append(df)
```

```python
df_list
```

```python
merged_df = reduce(lambda left, right: pd.merge(left, right, on=["시도명", "자격_대분류", "자격_코드명"], how="outer"), df_list)
merged_df
```

```python
merged_df.to_csv("C:/HANSSAK/SecureGate/Download/merged_df_250529_장기체류(등록+거소)외국인_데이터셋.csv", encoding='cp949', index=False)
```

```python
# 안전한 계산을 위해 0으로 나누는 경우 대비
def calc_rate_change(new, old):
    return (new - old) / old if old != 0 else None

# 증감률 계산
merged_df["20to21"] = (((merged_df["2021_외국인수"] - merged_df["2020_외국인수"]) / merged_df["2020_외국인수"]) * 100).round(2)
merged_df["21to22"] = (((merged_df["2022_외국인수"] - merged_df["2021_외국인수"]) / merged_df["2021_외국인수"]) * 100).round(2)
merged_df["22to23"] = (((merged_df["2023_외국인수"] - merged_df["2022_외국인수"]) / merged_df["2022_외국인수"]) * 100).round(2)
merged_df["23to24"] = (((merged_df["2024_외국인수"] - merged_df["2023_외국인수"]) / merged_df["2023_외국인수"]) * 100).round(2)

merged_df
```
