```python
title_list = []
date_list = []
writer_list = []
coauthor_list = []
content_list = []
kor_summary_list = []
eng_summary_list = []

# 드라이버 불러오기
driver = webdriver.Chrome()
# 페이지 열기
driver.get('https://www.mrtc.re.kr/data/08.php')
time.sleep(3)

# # 초기 element 값 설정
# brief_element = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[3]/table/tbody/tr[1]/td[3]/a')
# hrep = brief_element.get_attribute('href')

# 페이지 초기값
page_num = 3

# while문을 이용하여 전체 페이지 순회
while page_num <= 13:
    print(f'페이지 {page_num}')
    # for문을 이용하여 전체 페이지 문서 다운로드
    for i in range(10):
        brief_element = driver.find_elements(By.CLASS_NAME, 'subject > a')
        num = f"{page_num-4}{i+1}"
        # 총 문서 숫자 + 1
        p_num = 48-int(num)
        print(f'문서 {p_num}')
        hrep = brief_element[i].get_attribute('href')
        print(hrep)
        # 페이지 열기 
        driver.get(hrep)
        time.sleep(3)
        
        # 제목 가져오기
        title = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[1]/h3').text
        title_list.append(title)
        
        # 날짜 가져오기
        date = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/table/tbody/tr[1]/td').text
        date_list.append(date)
        
        # 저자 가져오기
        writer = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/table/tbody/tr[3]/td').text
        writer_list.append(writer)
        
        # 공동 저자 가져오기
        coauthor = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/table/tbody/tr[4]/td').text
        coauthor_list.append(coauthor)
        
        # 내용 가져오기
        content = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[3]').text
        content_list.append(content)
        
        # 한국어 요약본 정보
        try:
            kor_summary = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[2]/ul/li[2]/a').text
            kor_summary_list.append(kor_summary)
        except:
            kor_summary_list.append("없음")
        
        # 영어 오약본 정보
        try:
            eng_summary = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[2]/ul/li[3]/a').text
            eng_summary_list.append(eng_summary)
        except:
            eng_summary_list.append("없음")
        
        driver.back()
        time.sleep(3)
        
    # 다음 페이지로 이동
    next_page = driver.find_element(By.XPATH, f'//*[@id="inner_wrap"]/div[2]/div[4]/a[{page_num}]')
    next_page.click()
    time.sleep(3)
    
    page_num += 1
```

```python
# 이상한 문자들 제거
def sanitize_filename(filename):
    # 정규 표현식을 사용하여 특수 문자 제거
    sanitized = re.sub(r'[\\/:*?"<>|]', '', filename)
    sanitized = re.sub(r'[+·\-]', '', sanitized)  # +, ·, -를 공백으로 대체
    
    return sanitized
```
