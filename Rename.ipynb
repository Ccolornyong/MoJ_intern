```python
# # 원하는 파일 이름
# # git에서서 실행 시
file01_name = pd.read_json('1_research_report.json', encoding='utf-8')
file02_name = pd.read_json('2_policy_report.json', encoding='utf-8')
file03_name = pd.read_json('3_working_paper.json', encoding='utf-8')
file04_name = pd.read_json('4_issue_statistical_policy_brief.json', encoding='utf-8')
file05_name = pd.read_json('5_tasked_service_report.json', encoding='utf-8')
```

```python
file01_name.rename(columns={'kor_summery':'kor_summary', 'eng_summery':'eng_summary'}, inplace=True)
file02_name.rename(columns={'kor_summery':'kor_summary', 'eng_summery':'eng_summary'}, inplace=True)
file03_name.rename(columns={'kor_summery':'kor_summary', 'eng_summery':'eng_summary'}, inplace=True)
file04_name.rename(columns={'kor_summery':'kor_summary', 'eng_summery':'eng_summary'}, inplace=True)
file05_name.rename(columns={'kor_summery':'kor_summary', 'eng_summery':'eng_summary'}, inplace=True)
```

```python
# 각 리스트 2번째 페이지 크롤링 안됨

title_list = []
date_list = []
writer_list = []
coauthor_list = []
content_list = []
kor_summary_list = []
eng_summary_list = []

# 드라이버 불러오기
driver = webdriver.Chrome()
# 페이지 열기
driver.get('https://www.mrtc.re.kr/data/08.php')
time.sleep(3)

for i in range(10):
    subject_link = driver.find_elements(By.CLASS_NAME, 'subject > a')
    hrep = subject_link[i].get_attribute('href')
    driver.get(hrep)
    time.sleep(3)
    
    # 제목 가져오기
    title = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[1]/h3').text
    title_list.append(title)
    
    # 날짜 가져오기
    date = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/table/tbody/tr[1]/td').text
    date_list.append(date)
    
    # 저자 가져오기
    writer = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/table/tbody/tr[3]/td').text
    writer_list.append(writer)
    
    # 공동 저자 가져오기
    coauthor = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/table/tbody/tr[4]/td').text
    coauthor_list.append(coauthor)
    
    # 내용 가져오기
    content = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[3]').text
    content_list.append(content)
    
    # 한국어 요약본 정보
    try:
        kor_summary = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[2]/ul/li[2]/a').text
        kor_summary_list.append(kor_summary)
    except:
        kor_summary_list.append("없음")
    
    # 영어 오약본 정보
    try:
        eng_summary = driver.find_element(By.XPATH, '//*[@id="inner_wrap"]/div[2]/div[2]/div/div[2]/ul/li[3]/a').text
        eng_summary_list.append(eng_summary)
    except:
        eng_summary_list.append("없음")
    
    driver.back()
    time.sleep(3)
```

```python
import json

data_list = []

for i in range(len(title_list)):
    data = {
        'title': title_list[i],
        'date': date_list[i],
        'writer': writer_list[i],
        'coauthor': coauthor_list[i],
        'content': content_list[i],
        'kor_summary': kor_summary_list[i],
        'eng_summary': eng_summary_list[i]
    }
    data_list.append(data)
```
