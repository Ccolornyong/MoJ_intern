### 기존 final data에 2025 data 추가하기

```python
import pandas as pd
import numpy as np
```

```python
final_df = pd.read_csv("C:/HANSSAK/SecureGate/Download/final_20_24_250529_장기체류(등록+거소)외국인_데이터셋 (2).csv", encoding='cp949')
thisYear_df = pd.read_csv("C:/HANSSAK/SecureGate/Download/2025_250529_장기체류(등록+거소)외국인_데이터셋.csv", encoding='cp949')

thisYear_df
```

```python
thisYear_df['시도명'].unique()
```

```python
thisYear_df["시도명"] = thisYear_df["시도명"].replace("강원특별자치도", "강원도")
thisYear_df["시도명"] = thisYear_df["시도명"].replace("전북특별자치도", "전라북도")
```

```python
final_df['시도명'].unique()
```

```python
merged_df = final_df.merge(thisYear_df[['시도명', '자격_대분류', '2025_외국인수']],
                           on=['시도명', '자격_대분류'],
                           how='left')

final_df['2025_외국인수'] = merged_df['2025_외국인수']
```

```python
merged_df
```

### 24to25 추가

```python
# 안전한 계산을 위해 0으로 나누는 경우 대비
def calc_rate_change(new, old):
    return (new - old) / old if old != 0 else None

# 증감률 계산
merged_df["24to25"] = (((merged_df["2025_외국인수"] - merged_df["2024_외국인수"]) / merged_df["2024_외국인수"]) * 100).round(2)

merged_df
```
