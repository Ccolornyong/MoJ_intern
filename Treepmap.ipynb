### 트리맵 전처리 코드

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\HANSSAK\\SecureGate\\Download\\250609_장기체류(등록+거소)외국인_데이터셋(202504_05)_세종시....csv", encoding='cp949')
df
```

```python
df_202504 = df.loc[
    (df['기준년월'] == 202504) & (df['시도명'] == '경기도'),
    ['기준년월', '시도명', '시군구명', '대분류자격', '소분류자격', '국적', '외국인수']
]
df_202504
```

```python
df_202505 = df.loc[
    (df['기준년월'] == 202505) & (df['시도명'] == '경기도'),
    ['기준년월', '시도명', '시군구명', '대분류자격', '소분류자격', '국적', '외국인수']
]
df_202505
```

### 시군구명, 대분류자격 별로 외국인수 합치기

```python
df_202504_groupby = df_202504.groupby(
    ['시도명', '시군구명', '대분류자격'], as_index=False
)['외국인수'].sum()

df_202505_groupby = df_202505.groupby(
    ['시도명', '시군구명', '대분류자격'], as_index=False
)['외국인수'].sum()
```

```python
df_202504_groupby
```

```python
df_202505_groupby
```

### 이제 두 df 합치기

```python
Treemap_2025_경기도 = df_202504_groupby.merge(
    df_202505_groupby,
    on=['시도명', '시군구명', '대분류자격'],
    suffixes=('_202504', '_202505')
)

Treemap_2025_경기도.rename(columns={
    '외국인수_202504': '202504_외국인수',
    '외국인수_202505': '202505_외국인수',
    '대분류자격': '자격_대분류'
}, inplace=True)

Treemap_2504_2505_경기도 = Treemap_2025_경기도
```

```python
Treemap_2504_2505_경기도
```

### 자격_코드명 추가

```python
name = pd.read_excel("C:/HANSSAK/SecureGate/Download/250529_대분루자격코드명_메타/250529_대분류자격코드명_메타.xlsx")
name = dict(zip(name['자격_대분류'], name['LW_NM']))
```

```python
name
```

```python
Treemap_2504_2505_경기도['자격_코드명'] = Treemap_2504_2505_경기도['자격_대분류'].map(name)
Treemap_2504_2505_경기도
```

```python
# 안전한 계산을 위해 0으로 나누는 경우 대비
def calc_rate_change(new, old):
    return (new - old) / old if old != 0 else None

# 증감률 계산
Treemap_2504_2505_경기도["2504to2505"] = (((Treemap_2504_2505_경기도["202505_외국인수"] - Treemap_2504_2505_경기도["202504_외국인수"]) / Treemap_2504_2505_경기도["202504_외국인수"]) * 100).round(2)

Treemap_2504_2505_경기도
```

```python
Treemap_2504_2505_경기도.loc[Treemap_2504_2505_경기도['자격_대분류']=='00', '자격_코드명'] = '자격없음'
Treemap_2504_2505_경기도
```
